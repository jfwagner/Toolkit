# Introduction
We can conveniently describe a two dimensional magnetic field in terms of a complex field, with a [multipole expansion](https://www.wikiwand.com/en/Multipole_expansion):


\begin{equation*}
    \boxed{B_z + iB_x = \sum_n (b_n + i a_n)(x + i z)^n}
\end{equation*}

where $b_n$ is the **normal** component and $a_n$ the **skew** component.

* $n = 0 \ \ \rightarrow \ $ Dipole
* $n = 1 \ \ \rightarrow \ $ Quadrupole
* $n = 2 \ \ \rightarrow \ $ Sextupole
* $n = 3 \ \ \rightarrow \ $ Octupole

## Theory

Quadrupole magnets are used to provide a magnetic field whose magnitude grows rapidly with the radial distance from its longitudinal axis (see [field plot](#id-subsubsection121)). This is used for beam size control, e.g. to **focus and defocus the beam**.

<div id='id-subsubsection111'/>
### Field
For a generic quadrupole, we get:

\begin{equation*}
    \vec{B_z} = (b_1 x - a_1 z) \mathbf{\hat{z}} \ \ \ ; \ \ \ \vec{B_x} = (b_1 z + a_1 x) \mathbf{\hat{x}}
\end{equation*}

and for a normal (or upright) quadrupole, we have that $a_1 = 0$:

\begin{equation*}
    \vec{B_z} = b_1 x \mathbf{\hat{z}} \ \ \ ; \ \ \ \vec{B_x} = b_1 z \mathbf{\hat{x}}\ \ \Rightarrow \ \ \boxed{\vec{B_{quad}} = \vec{B_z} + \vec{B_x} = b_1 (x \mathbf{\hat{z}} + z \mathbf{\hat{x}})}
\end{equation*}

You can find more detailed calculus [here](http://www.lns.cornell.edu/~liepe/webpage/docs/P4456L5.pdf).

In the air space of the magnet which contains neither iron nor current conductors (i.e. no current densities or dynamic electric fields), [Ampère's Law](https://www.wikiwand.com/en/Amp%C3%A8re's_circuital_law) gives:

\begin{equation*}
    \nabla \times B = 0 \ \ \ \rightarrow \ \ \ \frac{\partial B_x}{\partial z} = \frac{\partial B_z}{\partial x}
\end{equation*}

Which is verified by our expressions of $B_z$ and $B_x$:

\begin{equation*}
    \frac{\partial B_z}{\partial x} = b_1 \ \ \ ; \ \ \ \frac{\partial B_x}{\partial z} = b_1 \ \Rightarrow \ \ \boxed{\vec{B_{quad}} = \frac{\partial B_z}{\partial x} (x \mathbf{\hat{z}} + z \mathbf{\hat{x}})}
\end{equation*}

The **sign of the gradient** $b_1$ determines wheter the quadrupole **focuses** or **defocuses** particles in the horizontal plane.

This also means that the field can be written as the gradient of a potential:

\begin{equation*}
    \vec{B} = \nabla \vec{V} \ \ \ ; \ \ \  V(x,z) = b_1 x z
\end{equation*}

The equipotential lines are the hyperbolas $xz = \text{constant}$. The field lines are perpendicular to them. If the relative permeability of the iron is large $\mu_r \gg 1$, iron pole shows with hyperbolic contour generate a rather pure quadrupole field.

This is why quadrupoles consist of four magnets (two nort, two south) with **hyperbolic contour**.

<img src="files/img/quad.png" width="20%" height="20%">

If we know consider the [force](https://www.wikiwand.com/en/Lorentz_force) due to the magnetic field:

\begin{equation*}
    \vec{F} = e(\vec{v} \times \vec{B}) =
    \left\{
        \begin{aligned}
        & F_x = e v B_z(x, z) = -e v b_1 x \\
        & F_z = -e v B_x(x, z) = e v b_1 z \\
        \end{aligned}
    \right.
\end{equation*}

We can see that a quadrupole **focuses in one plane**, while **defocusing on the other**. [Earnshaw's theorem
](https://www.wikiwand.com/en/Earnshaw's_theorem) shows that simultaneous focusing in two directions at once is impossible.


The horizontal force component depends only on the horizontal position of the particle trajectory (the same with the vertical coordinate). This means that for a linerar machine containing only dipoles and quadrupoles, the horizontal and vertical betatron oscillations are completely decoupled.

<div id='id-subsubsection112'/>
### Alternating gradient principle

If a focusing quadrupole and a defocusing quadrupole are placed immediately next to each other, their fields completely cancel out. But if there is a space between them (and the length of this has been correctly chosen), the overall effect is focusing in both horizontal and vertical planes. A lattice can then be built up enabling the transport of the beam over long distances - for example round an entire ring. 

A common lattice is a **FODO lattice** consisting of a basis of a focusing quadrupole, 'nothing' (often a bending magnet), a defocusing quadrupole and another length of 'nothing':

### Equations of motion

For a charged particle passing through the center of a quadrupole, the magnetic field and the Lorentz force are zero. At a displacement $(x, z)$ from the center, the Lorentz force for a particle of charge $e$ and velocity $v$ along the $\hat{s}$ direction is:

\begin{equation*}
    \vec{F} = e v \mathbf{\hat{s}} \times b_1 (x \mathbf{\hat{z}} + z \mathbf{\hat{x}}) = - e v b_1 z \mathbf{\hat{z}} + ev b_1 x \mathbf{\hat{x}}
\end{equation*}

Decoupling for the $x$ coordinate:

\begin{equation*}
    \vec{F_x} = m \ddot{x} = e v b_1 x \ \ \rightarrow \ \ m \frac{d^2x}{dt^2} = m \left( \frac{ds}{dt} \right)^2 \frac{d^2 x}{ds^2} = m v^2 x''
\end{equation*}

\begin{equation*}
    m v^2 x'' = e v b_1 x \ \ \rightarrow \ \ x'' - \frac{e}{p} b_1 x = 0
\end{equation*}

We obtain $\boxed{x'' - k_1 x = 0}$, and equivalently for the $z$ coordinate $\boxed{z'' + k_1 z = 0}$.

$\boxed{k_1 = \frac{e}{p} b_1}$, where the gradient has been normalized to the momentum of the particle, thus defining the **quadrupole strength**. 

Numerically:

\begin{equation*}
    k \left[ \text{m}^{-2}\right] = 0.2998 \frac{b_1 \text{[T/m]}}{p \text{[GeV/c]}}
\end{equation*}

For a constant $k_1$, we obtain the familiar **harmonic oscillator** equation, being the strength of the quadrupole the restoring force.

Let's consider the equation for the $z$ coordinate. The solution of any linear second order equation of the form of the harmonic oscillator is uniquely determined by the initial values of $z$ and $z'$:

\begin{equation*}
    z (s) = a z(s_0) + b z'(s_0) \ \ \ ; \ \ \ z' (s) = c z(s_0) + d z'(s_0)
\end{equation*}

A possible set of initial conditions that satisfy the equation is for example $ a = 1, \ b = 0 \ \ ; \ \ c = 0, \ d = 1$

In matrix notation this can be written as:

\begin{equation*}
  Z(s) =
  \begin{pmatrix}
    z(s) \\ z'(s)
  \end{pmatrix} =
  \begin{pmatrix}
    a & b \\
    c & d
  \end{pmatrix} 
  \begin{pmatrix}
    z(s_0) \\ z'(s_0)
  \end{pmatrix} 
\end{equation*}

The matrix formulation is useful because it separates the properties of the general solution from those due to a specific intial condition. The matrix depends only on $k_1(s)$ and the interval length $s-s_0$. 

The solutions to this equation are the *cosinelike* and *sinelike* trajectories:

<img src="files/img/sinelike.png" width="20%" height="20%">

\begin{equation}
  Z(s) =
  \begin{pmatrix}
    C & S \\
    C' & S'
  \end{pmatrix} 
  \begin{pmatrix}
    z(s_0) \\ z'(s_0)
  \end{pmatrix} 
\end{equation}


\begin{equation}
  C'' + k_1 C = 0 \ \ ; \ \ S'' + k_1 S = 0
\end{equation}

satisfying the initial conditions described above. 

We define $\mu = (s-s_0)\sqrt{|k_1|} = l\sqrt{|k_1|}$, $l$ being the length of the quadrupole magnet. We assume that the field is independent of $s$ inside the magnet and drops abruptly to zero at the ends (hard edge model). We get:

\begin{equation}
\begin{aligned}
 & \begin{pmatrix}
    C & S \\
    C' & S'
  \end{pmatrix} =
  \begin{pmatrix}
    \cos{\mu} & \frac{1}{\sqrt{|k_1|}}\sin{\mu} \\
    -\sqrt{|k_1|} \sin{\mu} &  \cos{\mu}
  \end{pmatrix} \ \ \ \ && \text{for} \ k_1 > 0 \ \text{(focusing)} \\
 & \begin{pmatrix}
    C & S \\
    C' & S'
  \end{pmatrix} =
  \begin{pmatrix}
    \cosh{\mu} & \frac{1}{\sqrt{|k_1|}}\sinh{\mu} \\
    -\sqrt{|k_1|} \sinh{\mu} &  \cosh{\mu}
  \end{pmatrix} \ \ \ \ && \text{for} \ k_1 < 0 \ \text{(defocusing)} \\
  & \begin{pmatrix}
    C & S \\
    C' & S'
  \end{pmatrix} =
  \begin{pmatrix}
   1 & l \\
   0 & 1
  \end{pmatrix} \ \ \ \ && \text{for} \ k_1 = 0 \ \text{(drift space)}
\end{aligned}
\end{equation}

### Group theory

We can observe that the quadrupole matrices have the form of a [fractional Fourier transform](https://www.wikiwand.com/en/Linear_canonical_transformation#/Special_cases), which corresponds to a **rotation by an arbitrary angle**. The drift matrix is a [Fresnel transform](https://www.wikiwand.com/en/Linear_canonical_transformation#/Special_cases), which corresponds to **shearing**. This implies several things:

* The effect of the quadrupoles and of the drift spaces represented by this matrices is a [linear canonical transformation](https://www.wikiwand.com/en/Linear_canonical_transformation). This means that this linear map preserves the symplectic structure, i.e. **preserves the area in phase space**. 
* This implies that the determinant of the matrix is always 1, making it part of the special linear group $\mathbb{SL}_2\mathbb{(R)}$ (group of all real 2 x 2 matrices with determinant one). It is the group of all linear transformations of $\mathbb{R}^2$ that preserve oriented area.
* The **quadrupole matrices** are called <a href="https://www.wikiwand.com/en/SL2(R)#Elliptic_elements">elliptic elements</a> of $\mathbb{SL}_2\mathbb{(R)}$ because $|tr(Z)| < 2$, which corresponds to a **rotation in phase space**.
* The **drift matrix** is called a <a href="https://www.wikiwand.com/en/SL2(R)#/Parabolic_elements">parabolic element</a> of $\mathbb{SL}_2\mathbb{(R)}$  because $|tr(Z)| = 2$ and corresponds to a **shear mapping in phase space**. This is why we need quadrupoles to focus.

### Transfer matrix method and stability condition
The matrix formulation we described above serves to apply the transfer matrix method. This method is used to analyze the propagation of electromagnetic waves through a
stratified medium. It is based on the fact that, according to Maxwell's equations, there
are simple continuity conditions for the electric field across boundaries from one medium
to the next. If the field is known at the beginning of the layer, the field at the end of
the layer can be derived from a simple matrix operation. A stack of layers can then be
represented as a system matrix, which is the product of the individual layer matrices. 

Translated to accelerator physics, we can obtain the effect of many different cells by multiplying their matrices:

<img src="files/img/matrix.png" width="30%" height="30%">

If we wanted to compute the effect of $N$ identical cells through $k$ revolutions it would simply be $[M(s)]^{Nk}$.

In order for the motion to be stable all the elements of the matrix $[M(s)]^{Nk}$ must remain bounded as $k$ increases indefinitely. To obtain the stability condition for this, we consider the eigenvalues of the matrix $M(s)= \begin{pmatrix}a & b \\ c & d\end{pmatrix}$:

\begin{equation*}
  Z(s) =
  \begin{pmatrix}
    z(s) \\ z'(s)
  \end{pmatrix} =
  \begin{pmatrix}
    a & b \\
    c & d
  \end{pmatrix} 
  \begin{pmatrix}
    z(s_0) \\ z'(s_0)
  \end{pmatrix} = {\lambda}_1 \ldotp z(s_0) +  {\lambda}_2 \ldotp  z'(s_0)
\end{equation*}

Since we have seen that this matrices are linear canonical transformations, we can apply the condition that the determinant must be equal to 1 $Det(M) = 1 \ \ \ \rightarrow \ \ \ {\lambda}_1 \ldotp {\lambda}_2 = 1$.


We could then generally write ${\lambda}_1 = e^{i\mu} \ \ \ ; \ \ \ {\lambda}_2 =  e^{-i\mu}$.

The trace will then be $Tr(M) = {\lambda}_1 + {\lambda}_2 = e^{i\mu} + e^{-i\mu} = 2 \cos{\mu}$.

For the movement to be bounded, we need $\mu$ to be real. If it were complex we would obtain an hyperbolic cosine which is not bounded, whereas $|\cos{\mu}| < 1$. The condition for stability is then:

\begin{equation*}
  \boxed{|Tr(M)|< 2 \ \ \ \leftrightarrow \ \ \  \mu \in \mathbb{R}}
\end{equation*}

If we remember the transfer matrix for a quadrupole, we see that $|Tr(M)|< 2$ corresponds to an elliptic element of $\mathbb{SL}_2\mathbb{(R)}$.

### Thin lens approximation

Since magnets act like lenses, we can also define a focal length $f$ for quadrupoles, i.e. how strongly the system converges or diverges light $ \int_0^lk_1(s)ds = \frac{1}{f} $.

<img src="files/img/focal.png" width="10%" height="10%">

In the thin lens limit the deflection is concentrated in a centered element with zero length: $ \text{lim} \ l \rightarrow 0 \ \ \ \ \text{keeping} \ \ \ k_1\ldotp l = \frac{1}{f} = const.$

In this approximation the quadrupole transfer matrices simplify to:

\begin{equation}
  \begin{pmatrix}
    1 & 0 \\
    -\frac{1}{f} & 1
  \end{pmatrix} \ \ \ \text{focusing} \ \ \ ; \ \ \
  \begin{pmatrix}
    1 & 0 \\
    \frac{1}{f} & 1
  \end{pmatrix} \ \ \  \text{defocusing}
\end{equation}

**Example: FODO lattice**

\begin{equation*}
    M =
    \begin{pmatrix}
        1 & L \\
        0 & 1
    \end{pmatrix}
    \begin{pmatrix}
        1 & 0 \\
        1/f & 1
    \end{pmatrix}
    \begin{pmatrix}
        1 & L \\
        0 & 1
    \end{pmatrix}
    \begin{pmatrix}
        1 & 0 \\
        -1/f & 1
    \end{pmatrix} =
    \begin{pmatrix}
        1 - L/f -\left(L/f\right)^2 & 2L + L^2/f \\
        -L/f^2 & 1 + L/f
    \end{pmatrix}
\end{equation*}

### Effects of quadrupole magnets on the particles

```python
# Creating the transverse distribution
particles = 6400

beta_x = 6.8
em_x = 1e-6

sigma_x = np.sqrt(beta_x*em_x)
sigma_xp = np.sqrt(em_x/beta_x)

print 'sigma x', sigma_x
print 'sigma xp', sigma_xp

ran_x = []
ran_xp = []
for i in range(0, particles):
	ran_x.append(gauss(0,1))
	ran_xp.append(gauss(0,1))

x_data = []
for particle_x in ran_x:
    x_data.append(float(particle_x*sigma_x))
xp_data = []
for particle_xp in ran_xp:
    xp_data.append(float(particle_xp*sigma_xp))

x = np.asarray(x_data)
xp = np.asarray(xp_data)

plt.scatter(x, xp)
plt.title('Initial distribution')
plt.ylabel("$x'$")
plt.xlabel('x')
plt.xlim([-0.01, 0.01])
plt.ylim([-0.01, 0.01])
plt.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
plt.ticklabel_format(style='sci',axis='x',scilimits=(0,0))
plt.grid(b=None, which='major')
```

```python
# FOCUSING QUADRUPOLE
l = 1.9  # m
k_1 = 0.5  # m^-2
f = 1 / (k_1*l)

# Not stable motion with these parameters, but useful to see the effect

a = np.divide(x,f)

x_f = x
xp_f = xp - a

plt.scatter(x_f, xp_f)
plt.xlim([-0.01, 0.01])
plt.ylim([-0.01, 0.01])
plt.ticklabel_format(style='sci',axis='x',scilimits=(0,0))
plt.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
plt.xlabel('x [m]')
plt.ylabel('xp [m]')
plt.grid(b=None, which='major')
```

```python
# DEFOCUSING QUADRUPOLE
x_d = x
xp_d = xp + a

plt.scatter(x_d, xp_d)
plt.xlim([-0.01, 0.01])
plt.ylim([-0.01, 0.01])
plt.ticklabel_format(style='sci',axis='x',scilimits=(0,0))
plt.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
plt.xlabel('x [m]')
plt.ylabel('xp [m]')
plt.grid(b=None, which='major')
```

```python
# FODO CELL

# Initialize matrices
focusing = np.zeros((2, 2))
defocusing = np.zeros((2, 2))
drift = np.zeros((2, 2))

# Fill in matrices
focusing[0,0] = 1
focusing[0,1] = 0
focusing[1,0] = -k_1
focusing[1,1] = 1

defocusing[0,0] = 1
defocusing[0,1] = 0
defocusing[1,0] = k_1
defocusing[1,1] = 1

drift[0,0] = 1
drift[0,1] = l
drift[1,0] = 0
drift[1,1] = 1

print 'focusing'
print  focusing
print ' '
print 'defocusing'
print defocusing
print ' '
print 'drift'
print drift
```

```python
# Multiplication of the matrices
total_matrix = dot(dot(dot(drift, defocusing), drift), focusing)
print total_matrix
```

```python
# Tracking transverse

particles = len(x)
turns = 1000
vector_transverse = np.vstack((x, xp))

x_turn = np.zeros((turns, particles))
xp_turn = np.zeros((turns, particles))

for t in xrange(0,turns):
    x_new = dot(total_matrix, vector_transverse)
    vector_transverse = x_new
    x_final = vector_transverse
    for i in xrange(0, particles):
        x_turn[t][i] = x_final[0][i]
        xp_turn[t][i] = x_final[1][i]

plt.scatter(x_final[0], x_final[1])
plt.title('After FODO')
plt.ylabel("$x'$")
plt.xlabel('x')
plt.xlim([-0.01, 0.01])
plt.ylim([-0.01, 0.01])
plt.ticklabel_format(style='sci',axis='y',scilimits=(0,0))
plt.ticklabel_format(style='sci',axis='x',scilimits=(0,0))
plt.grid(b=None, which='major')
```
